```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Data
df = pd.read_csv('/content/Crop Production data.csv')  # Assuming your data is in a CSV file
```

```python
# Handle missing values (if any)
df.dropna(subset=['Area', 'Production'], inplace=True)
```

```python
# Convert to correct data types
df['Crop_Year'] = pd.to_numeric(df['Crop_Year'])
df['Area'] = pd.to_numeric(df['Area'])
df['Production'] = pd.to_numeric(df['Production'])
```

```python
# 1. Overall Production Trends
total_production_by_year = df.groupby('Crop_Year')['Production'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(y=total_production_by_year.index, x=total_production_by_year.values)
plt.title('Total Crop Production in Andhra Pradesh Over Time')
plt.ylabel('Year.Month')
plt.xlabel('Total Production (Tonnes)')
plt.grid(axis='y')
plt.show()
```

```python

# 2. Production by State
top_states = df.groupby('State_Name')['Production'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_states.index, y=top_states.values)
plt.title('Top 10 Crop Producing States')
plt.xlabel('State')
plt.ylabel('Total Production (Tonnes)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Crop-Specific Analysis (Example: Rice)
rice_production = df[df['Crop'] == 'Rice'].groupby('Crop_Year')['Production'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=rice_production.index, y=rice_production.values)
plt.title('Rice Production in India Over Time')
plt.xlabel('Year')
plt.ylabel('Rice Production (Tonnes)')
plt.grid(axis='y')
plt.show()
```

```python
# 4. Seasonal Variations
production_by_season = df.groupby('Season')['Production'].sum()
plt.figure(figsize=(12, 6))
plt.pie(production_by_season, labels=production_by_season.index, autopct='%1.1f%%', startangle=140)
plt.title('Crop Production by Season')
plt.show()
```
