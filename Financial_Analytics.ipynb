```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
df = pd.read_csv('/content/Financial Analytics data.csv')  # Assuming your data is in a CSV file

# Handle missing values (if any)
df.dropna(subset=['Mar Cap - Crore', 'Sales Qtr - Crore'], inplace=True)

# Convert to correct data types
df['Mar Cap - Crore'] = pd.to_numeric(df['Mar Cap - Crore'])
df['Sales Qtr - Crore'] = pd.to_numeric(df['Sales Qtr - Crore'])
```

```python
# 1. Market Capitalization Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Mar Cap - Crore'], kde=True)
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Cap (Crore)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Quarterly Sales Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales Qtr - Crore'], kde=True)
plt.title('Distribution of Quarterly Sales')
plt.xlabel('Quarterly Sales (Crore)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Relationship Between Market Cap and Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=df)
plt.title('Relationship Between Market Cap and Quarterly Sales')
plt.xlabel('Market Cap (Crore)')
plt.ylabel('Quarterly Sales (Crore)')
plt.show()
```
